```python
def test(country_name):
  URL = ('http://apis.data.go.kr/1262000/CountryGnrlInfoService2/'
      'getCountryGnrlInfoList2?'
      'serviceKey=N6ub5E6R16ZbABSO%2FCmyepODW7qzhioaCLMJr1YgXaY7YkkJQlWArUPeE2aNJcI2Uxv82Bqq%2BmCqLkwvTPgS0w%3D%3D'
      '&returnType=json'
      '&pageNo=1'
      '&numOfRows=10'
      '&cond[country_nm::EQ]='+country_name)
  
  response = requests.get(URL)
  r = response.json()
  data = r['data']

  country_nm = [item['country_nm'] for item in data]
  country_lc = [item['country_lc'] for item in data]
  lang_nm = [item['lang_nm'] for item in data]
  country_cptl_nm = [item['country_cptl_nm'] for item in data]
  religion_cn = [item['religion_cn'] for item in data]
  climate_cn = [item['climate_cn'] for item in data]

  country_nm = pd.Series(country_nm)
  country_lc = pd.Series(country_lc)
  lang_nm = pd.Series(lang_nm)
  country_cptl_nm = pd.Series(country_cptl_nm)
  religion_cn = pd.Series(religion_cn)
  climate_cn = pd.Series(climate_cn)

  df = pd.DataFrame({'country':country_nm, 'country location': country_lc, 'language': lang_nm, 'capital': country_cptl_nm, 'religion': religion_cn, 'climate':climate_cn})
  return df
```

```python
test('감비아')
```

```python
iface_test = gr.Interface(test,gr.inputs.Dropdown(['가나','가봉','가이아나','감비아']),outputs=['dataframe'])
iface_test.launch(debug=True)
```
